```python
def recursive_fibonacci(n):
    if n <= 0:
        return 0
    if n == 1 or n == 2:
        return 1
    return recursive_fibonacci(n - 1) + recursive_fibonacci(n - 2)

def iterative_fibonacci(n):
    if n <= 0:
        return 0
    if n == 1 or n == 2:
        return 1
    a = 1
    b = 1
    for _ in range(3, n + 1):
        c = a+b
        b = a
        a = c
    return c
```

```python
import time

import matplotlib.pyplot as plt

# Measure execution time for different n values
n_values = [20, 30, 40, 45]
recursive_times = []
iterative_times = []

for n in n_values:
    # Measure recursive Fibonacci time
    start_time = time.time()
    recursive_fibonacci(n)
    recursive_times.append(time.time() - start_time)
    
    # Measure iterative Fibonacci time
    start_time = time.time()
    iterative_fibonacci(n)
    iterative_times.append(time.time() - start_time)

# Plot the results
plt.plot(n_values, recursive_times, label='Recursive Fibonacci', marker='o')
plt.plot(n_values, iterative_times, label='Iterative Fibonacci', marker='o')
plt.xlabel('n')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Fibonacci Algorithms')
plt.legend()
plt.grid()
plt.show()

# Time complexity:
# Recursive Fibonacci: Exponential time complexity, O(2^n)
# Iterative Fibonacci: Linear time complexity, O(n)
```
